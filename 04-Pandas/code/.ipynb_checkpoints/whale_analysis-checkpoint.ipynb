## S&P TSX 60 Returns

Read the S&P TSX 60 historic closing prices and create a new daily returns DataFrame from the data. 

```python
# ReReading S&P TSX 60 Closing Prices
sptsx = pd.read_csv(sptsx_csvpath, index_col="Date", parse_dates = True, infer_datetime_format=True).sort_index()

# Display sample data
sptsx.head()
```

```python
# Check Data Types
sptsx.dtypes
```

```python
# Create a working df
sptsx_df = sptsx.sort_index().copy()

# Fix Data Types
sptsx_df['Close'] = sptsx_df['Close'].str.replace('$', '')
sptsx_df['Close'] = sptsx_df['Close'].str.replace(',', '')
sptsx_df['Close'] = pd.to_numeric(sptsx_df['Close'])

# Display sample data
sptsx_df.head()
```

```python
# Convert Date to Index
#sptsx_df = sptsx_df.set_index(pd.to_datetime(sptsx_df["Date"],infer_datetime_format=True)).drop(columns=["Date"]).copy()

# Ensure dates are in sequential order
sptsx_df.sort_index(axis="rows", ascending=True, inplace=True)

# Display sample data
sptsx_df.head()
```

```python
# Calculate Daily Returns
# Drop nulls
sptsx_returns_df = sptsx_df.pct_change().dropna().copy()

#Shorten the column names
sptsx_returns_df.columns = ["TSX"]

# Display sample data
sptsx_returns_df.head()
```

```python
# Explore the whale data further
sptsx_returns_df.describe()
```

### Calculate Standard Deviations

```python
# Calculate the daily standard deviations of all portfolios
stddev_returns = pd.DataFrame(daily_returns.std(), columns = ['Std_Dev'])

# Display Data sorted fram Lowest to Highest Risk
stddev_returns.sort_values("Std_Dev")
```

### Determine which portfolios are riskier than the S&P TSX 60

```python
# Calculate  the daily standard deviation of S&P TSX 60
TSX_sd = stddev_returns.at["TSX", "Std_Dev"]
print(f'The daily standard deviation of S&P TSX 60 is: {TSX_sd}')
```

```python
# Determine which portfolios are riskier than the S&P TSX 60
print(f'The portfolios riskier than the S&P TSX 60 are: ')

# Display Std Dev Lowest to Highest - Least to Most Risky after the TSX 60
stddev_returns[stddev_returns["Std_Dev"] > (TSX_sd)].sort_values(by = "Std_Dev")
```

### Calculate the Annualized Standard Deviation

```python
# Calculate the annualized standard deviation (252 trading days)
annualized_std = stddev_returns * np.sqrt(252)

# Display Std Dev Lowest to Highest
annualized_std.sort_values(by = "Std_Dev")
```

## Rolling Statistics

Risk changes over time. Analyze the rolling statistics for Risk and Beta. 

1. Calculate and plot the rolling standard deviation for for all portfolios using a 21-day window.
2. Calculate the correlation between each stock to determine which portfolios may mimick the S&P TSX 60.
3. Choose one portfolio, then calculate and plot the 60-day rolling beta for it and the S&P TSX 60.

### Calculate and plot rolling `std` for all portfolios with 21-day window

```python
# Calculate the rolling standard deviation for all portfolios using a 21-day window
rolling_std_21 = daily_returns.rolling(21).std().dropna().copy()

# Display sample data
rolling_std_21.head()
```

```python
# Plot the rolling standard deviation
fig = plt.figure(figsize=(20,10))

plt.plot(rolling_std_21)
plt.title("Portfolio Rolling 21-Day Standard Deviation", fontsize = 20)
plt.ylabel("Standard Deviation", fontsize = 16)
plt.xlabel("Date", fontsize = 16)
plt.legend(rolling_std_21.columns)

plt.show()
```

### Calculate and plot the correlation

```python
# Calculate the correlation
correlation = daily_returns.corr()

# Display the correlation matrix
correlation
```

```python
# Correlation between different variables
corr = daily_returns.corr()

# Set up the matplotlib plot configuration
f, ax = plt.subplots(figsize=(12, 10))

# Generate a mask for upper traingle
mask = np.triu(np.ones_like(corr, dtype=bool))

# Configure a custom diverging colormap
cmap = sns.diverging_palette(230, 20, as_cmap=True)

# Draw the heatmap
sns.heatmap(corr, annot=True, mask = mask, cmap=cmap)

# Reference: https://vitalflux.com/correlation-heatmap-with-seaborn-pandas/
```

### Calculate and Plot Beta for a chosen portfolio and the S&P 60 TSX

```python
# Calculate covariance of a single portfolio
soros_covariance = daily_returns["SOROS"].cov(daily_returns["TSX"])
print(f'Soros Covariance is: {soros_covariance}')

# Calculate variance of S&P TSX
tsx_variance = daily_returns["TSX"].var()
print(f'TSX 60 variance is: {tsx_variance}')

# Computing beta
soros_beta = round((soros_covariance/tsx_variance),4)
print(f'Soros Beta is: {soros_beta}')
```

```python
# Create a rolling 21-Day Beta data set
soros_covariance_roll_21 = daily_returns["SOROS"].rolling(21).cov(daily_returns["TSX"]).copy()
tsx_variance_roll_21 = daily_returns["TSX"].rolling(21).var().copy()
soros_beta_roll_21 = pd.DataFrame(round((soros_covariance_roll_21/tsx_variance_roll_21),4),columns = ["Beta"]).dropna().sort_index()

# Display sample data
soros_beta_roll_21.head()
```

```python
# Plot beta trend
fig = plt.figure(figsize=(20,10))

plt.plot(soros_beta_roll_21)
plt.title("Rolling 21-Day Beta: Soros vs. TSX", fontsize = 20)
plt.ylabel("Beta", fontsize = 16)
plt.xlabel("Date", fontsize = 16)
plt.legend(soros_beta_roll_21.columns)

plt.show()
```

## Rolling Statistics Challenge: Exponentially Weighted Average 

An alternative way to calculate a rolling window is to take the exponentially weighted moving average. This is like a moving window average, but it assigns greater importance to more recent observations. Try calculating the [`ewm`](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.ewm.html) with a 21-day half life for each portfolio, using standard deviation (`std`) as the metric of interest.

```python
# Use `ewm` to calculate the rolling window
# FYI - column_appended_data.ewm(min_periods = 21, halflife = 21).mean().plot(figsize=(20,10))

daily_returns_ewm = daily_returns.ewm(min_periods = 21, halflife = 21).std().copy()

fig = plt.figure(figsize=(20,10))

plt.plot(daily_returns_ewm)
plt.title("Portfolio EWM Standard Deviation", fontsize = 20)
plt.ylabel("Standard Deviation", fontsize = 16)
plt.xlabel("Date", fontsize = 16)
plt.legend(daily_returns_ewm.columns)

plt.show()
```

# Sharpe Ratios
In reality, investment managers and thier institutional investors look at the ratio of return-to-risk, and not just returns alone. After all, if you could invest in one of two portfolios, and each offered the same 10% return, yet one offered lower risk, you'd take that one, right?

### Using the daily returns, calculate and visualize the Sharpe ratios using a bar plot

```python
# Annualized Sharpe Ratios
sharpe_ratios = pd.DataFrame((daily_returns.mean() * 252) / (daily_returns.std() * np.sqrt(252)), columns = ['Sharpe']).sort_values(by = "Sharpe")

# Display Sharpe Ratios ordered by Lowest to Highest Risk Adjusted Returns
sharpe_ratios.sort_values(["Sharpe"])
```

```python
# Visualize the sharpe ratios as a bar plot
sharpe_ratios.plot(kind = "bar", title = "Portfolio Sharpe Ratios")
```

## Choose 3-5 custom stocks with at last 1 year's worth of historic prices and create a DataFrame of the closing prices and dates for each stock.

```python
# Reading data from 1st stock
l_df = pd.read_csv(l_csvpath)

# Display sample data
l_df.head()
```

```python
# Re-Read the the csv file coveting the date colum to the index.
l_df = pd.read_csv(l_csvpath, index_col="Date", parse_dates = True, infer_datetime_format=True)
l_df = l_df.drop("Symbol", axis = 1)

#Shorten the column names
l_df.columns = ["L"]

# Display sample data
l_df.head()
```

```python
# Reading data from 2nd stock
# Re-Read the the csv file coveting the date colum to the index.
otex_df = pd.read_csv(otex_csvpath, index_col="Date", parse_dates = True, infer_datetime_format=True)
otex_df = otex_df.drop("Symbol", axis = 1)

#Shorten the column names
otex_df.columns = ["OTEX"]

# Display sample data
otex_df.head()
```

```python
# Reading data from 3rd stock
# Re-Read the the csv file coveting the date colum to the index.
shop_df = pd.read_csv(shop_csvpath, index_col="Date", parse_dates = True, infer_datetime_format=True)
shop_df = shop_df.drop("Symbol", axis = 1)

#Shorten the column names
shop_df.columns = ["SHOP"]

# Display sample data
shop_df.head()
```

```python
# Combine all stocks in a single DataFrame
daily_close = pd.concat([l_df, otex_df, shop_df], axis="columns", join="inner")

# Display sample data
daily_close.head()
```

```python
# Reset Date index
daily_close.sort_index(inplace = True)

# Display sample data
daily_close.head()
```

```python
# Reorganize portfolio data by having a column per symbol

# Done while loading the stocks
```

```python
# Calculate daily returns
daily_close["L_Return"] = daily_close["L"].pct_change()
daily_close["OTEX_Return"] = daily_close["OTEX"].pct_change()
daily_close["SHOP_Return"] = daily_close["SHOP"].pct_change()

# Create "Daily Return" df; Drop "Closing Price" columns; and Drop NAs
daily_return = daily_close.drop(columns = ["L", "OTEX", "SHOP"]).dropna().copy()

# Display sample data
daily_return.head()
```

## Calculate the weighted returns for the portfolio assuming an equal number of shares for each stock

```python
# Set weights
weights = [1/3, 1/3, 1/3]

# Calculate portfolio return
portfolio_return = pd.DataFrame(daily_return.dot(weights), columns = ['Portfolio'])

# Display sample data
portfolio_return.sample(10)
```

## Join your portfolio returns to the DataFrame that contains all of the portfolio returns

```python
# Join your returns DataFrame to the original returns DataFrame
all_portfolios = pd.concat([portfolio_return, daily_returns], axis="columns", join="inner").sort_index()

# Display sample data
all_portfolios.head(10)
```

```python
# Only compare dates where return data exists for all the stocks (drop NaNs)
all_portfolios.dropna()

# Display sample data
all_portfolios.head()
```

## Re-run the risk analysis with your portfolio to see how it compares to the others

```python
# Calculate and plot Beta

# Calculate covariance of a single portfolio
portfolio_covariance_roll_21 = all_portfolios["Portfolio"].rolling(60).cov(all_portfolios["TSX"])

# Calculate variance of S&P TSX
tsx_60_variance_roll_21 = all_portfolios["TSX"].rolling(60).var()

# Computing beta
portfolio_beta_roll_21 = round((portfolio_covariance_roll_21/tsx_60_variance_roll_21),4)
```
