## Import Libraries

```python
# -*- coding: UTF-8 -*-
"""PyRamen Homework Starter."""

# @TODO: Import libraries
import csv
from pathlib import Path
```

## Set Data File Pathes

```python
# @TODO: Set file paths for menu_data.csv and sales_data.csv
menu_filepath = Path('Data/menu_data.csv')
sales_filepath = Path('Data/sales_data.csv')
```

## Read Menu Data File

```python
# @TODO: Initialize list objects to hold our menu and sales data
menu = []

# @TODO: Read in the menu data into the menu list
with open(menu_filepath, 'r') as csvfile:
    # creating a csv reader object
    csvreader = csv.reader(csvfile)
     
    # extracting field names through first row --> Add Field Names to a List
    menu_fields = next(csvreader)
 
    # extracting each data row one by one --> Create a List of Rows where each Row is a List of Values
    for row in csvreader:
        menu.append(row)


# Display Menu Data Set 
#print(menu_fields)
#print('------------')
#for item in menu:
#    print(item)
```

## Read Sales Data File

```python
# @TODO: Initialize list objects to hold our menu and sales data
sales = []

# @TODO: Read in the sales data into the sales list
with open(sales_filepath, 'r') as csvfile:
    # creating a csv reader object
    csvreader = csv.reader(csvfile)
     
    # extracting field names through first row --> Add Field Names to a List
    sales_fields = next(csvreader)
 
    # extracting each data row one by one --> Create a List of Rows where each Row is a List of Values
    for row in csvreader:
        sales.append(row)

# Display Sales Data - frist 10 rows
#print(sales_fields)
#print('------------')
#for sale in range(10):
#    print(sales[sale])
```

## Main Code Block: Populate Report with Menu Item Finacial Metrics

```python
# @TODO: Initialize dict object to hold our key-value pairs of items and metrics
report = {}

# Initialize a row counter variable
row_count = 0

# @TODO: Loop over every row in the sales list object

for sale in sales:


    # Line_Item_ID,Date,Credit_Card_Number,Quantity,Menu_Item
    # @TODO: Initialize sales data variables
    sales_line = sale[0]
    sales_date = sale[1]
    sales_creditcard = sale[2]
    sales_qualtity = int(sale[3])
    sales_item = sale[4] 
    #print(sales_item + ' --- ' + str(sales_qualtity))

    # @TODO:
    # If the item value not in the report, add it as a new entry with initialized metrics
    # Naming convention allows the keys to be ordered in logical fashion, count, revenue, cost, profit
    if sales_item not in report.keys():
        report.update({sales_item:{"01-Count" : 0, "02-Revenue" : 0, "03-COGS" : 0, "04-Profit" : 0  }})


    # @TODO: For every row in our sales data, loop over the menu records to determine a match
    for item in menu:

        # Item,Category,Description,Price,Cost
        # @TODO: Initialize menu data variables
        menu_item = item[0]
        item_category = item[1]
        item_description = item[2]
        item_price = float(item[3])
        item_cost = float(item[4])
        item_profit = item_price - item_cost
        #print(menu_item + ' --- ' + str(sales_qualtity)+ ' --- ' + str(item_price)+ ' --- ' + str(item_cost)+ ' --- ' + str(item_profit))
        
        
        # @TODO: Calculate profit of each item in the menu data
        count = sales_qualtity
        revenue = float(sales_qualtity) * item_price
        cogs = float(sales_qualtity) * item_cost
        profit = float(sales_qualtity) * item_profit

        # @TODO: If the item value in our sales data is equal to the any of the items in the menu, then begin tracking metrics for that item
        if sales_item == menu_item:

            # @TODO: Print out matching menu data
            if row_count < 10:     # --> Print only 1st 10 rows to save time
                print(sales_item + ': Quantity: ' + str(sales_qualtity)+ ' Unit Price ' + str(item_price)+ ' Unit Cost: ' + str(item_cost)+ ' Unit Profit: ' + str(item_profit)+ ' Revenue: ' + str(revenue)+ ' COGS' + str(cogs) + ' Profit: ' + str(profit)) 





            # @TODO: Cumulatively add up the metrics for each item key
            report[sales_item]["01-Count"] += sales_qualtity
            report[sales_item]["02-Revenue"] += revenue
            report[sales_item]["03-COGS"] += cogs
            report[sales_item]["04-Profit"] += profit
            #print(str(report[sales_item]["01-Count"])+ ' --- ' + str(report[sales_item]["02-Revenue"])+ ' --- ' + str(report[sales_item]["03-COGS"])+ ' --- ' + str(report[sales_item]["04-Profit"]))
            



        # @TODO: Else, the sales item does not equal any fo the item in the menu data, therefore no match
        else: 
            if row_count < 10:      # --> Print only 1st 10 rows to save time
                print(f"{sales_item} does not equal {menu_item}! NO MATCH!")


    # @TODO: Increment the row counter by 1
    row_count += 1
```

## Print Number of Sales Records

```python
# @TODO: Print total number of records in sales data
print(f"Number of Sales Records: {row_count}")
```

## Write the Report Dictionary to txt File: FinancialAnalysis_PyRamen.txt

```python
# @TODO: Write out report to a text file (won't appear on the command line output)
with open("FinancialAnalysis_PyRamen.txt", mode = "w") as f:
    for menu_item in report.items():
        f.write("('" + menu_item[0] + "', {'01-Count': " + str(menu_item[1]["01-Count"])+ ", '02-Revenue': " + str(menu_item[1]["02-Revenue"])+ ", '03-COGS': " + str(menu_item[1]["03-COGS"])+ ", '04-Profit': " + str(menu_item[1]["04-Profit"]) + "}) \n")
            
```
